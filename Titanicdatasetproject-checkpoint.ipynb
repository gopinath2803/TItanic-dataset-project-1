```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
     # load data
df=sns.load_dataset("titanic")
print(df)
df.columns
sns.countplot(df['class'])
df['class'].value_counts()
     # set up the category plots
g=sns.catplot(x="class",y="survived",hue="sex",data=df,kind="bar")
g=sns.catplot(x="class",y="survived",hue="sex",data=df,kind="violin")
     # show plot
plt.show()
sns.heatmap(df.isnull())
```
